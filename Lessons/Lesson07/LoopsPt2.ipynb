#### 1.0 _for_ loop

Last week we discussed a **_while_** loop.  A **_while_** loop uses a condition to determine whether or not we enter the loop.  If the condition is True we enter, when the condition is False we exit the loop.  So in a **_while_** we do something over and over again until a condition becomes False.

A **_while_** loop was ideal for looping over until a conditional turned to False, but what if I had a data structure and wanted to loop over all the elements in the data structure?

This is what the **_for_** loop is ideally suited for.  In a **_for_** loop we loop over every item in a data structure.  

The structure of the **_for_** is like this:

```
for <loop_variable_name> in <data_structure>:
    <loop_body>
```

The **data_structure** above refers to the data structure that we are going to loop over.  The **loop_variable_name** is just a variable that will be assigned the current element from the data structure.  Every iteration of the loop the value assigned to the loop variable will change.

##### 1.1 Examples

The final thing we are going to talk about with loops is the builtin function called **_range_**
The range function allows us to create a sequence of numbers up to a specified limit.  The **_range_** function takes one argument, the upper limit.  The sequence generated by **_range_** goes from 0 to 1 minus the upper limit.

````python
range(10)
````
